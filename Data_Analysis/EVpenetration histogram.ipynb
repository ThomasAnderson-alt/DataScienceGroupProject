```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 加载数据
df = pd.read_csv('Master_Panel_Data_Final.csv')

# 2. 筛选 2025 年的地方政府（LAD）级别数据，以获得最细致的地理分布
df_2025 = df[(df['Geo_level'] == 'LAD') & (df['year'] == 2025)].copy()
df_2025 = df_2025.dropna(subset=['ev_penetration'])

# 3. 绘制直方图 (Histogram) 和 核密度曲线 (KDE)
plt.figure(figsize=(10, 6))
sns.histplot(df_2025['ev_penetration'], kde=True, color='teal', bins=30)

# 添加中位数和均值线，更直观展示偏态
mean_val = df_2025['ev_penetration'].mean()
median_val = df_2025['ev_penetration'].median()
plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}%')
plt.axvline(median_val, color='blue', linestyle='-', label=f'Median: {median_val:.2f}%')

plt.title('Distribution of EV Penetration across UK Local Authorities (2025)', fontsize=14)
plt.xlabel('EV Penetration Rate (%)', fontsize=12)
plt.ylabel('Number of Authorities', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)

# 打印偏度 (Skewness)
print(f"Skewness: {df_2025['ev_penetration'].skew():.2f}")
```
